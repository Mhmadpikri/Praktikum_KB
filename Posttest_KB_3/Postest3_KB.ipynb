```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### tampilkan data
#### data cuaca setiap jam

```python
df = pd.read_csv('Weather.csv')
df
```

### MENAMPILKAN JUMLAH DATA PADA COLOM CONDS

```python
df['conds'].value_counts()
```

### MEMBUAT GRAFIK BATANG
##### Disini saya menampilkan kolim icon dan rain dimana hasilnya seperti dibawah

```python
# Dengan Matplotlib

x = list(df["rain"])
y = list(df["icon"])

plt.xlabel("cuaca rain")
plt.ylabel("keadaan sekitar")
plt.title("Grafik Batang")
plt.bar(x,y,color="green")
plt.show()
```

### MENAMPILKAN GRAFIK GARIS

```python
x = list(df["snow"])
y = list(df["icon"])
plt.title("Grafik Garis")
plt.xlabel("keadaan sekitar")
plt.ylabel("keadaan bersalju")
plt.plot(x,y)
plt.show()
```

### MENAMPILKAN GRAFIK SCATTER PLOT

```python
#Dengan Matplotlib
plt.figure(figsize=(10,5))
plt.scatter(x=df["rain"],y=df["conds"])
plt.xlabel("cuaca hujan")
plt.ylabel("kondisi")
plt.show()
```

### MENAMPILKAN GRAFIK PIE
##### disini saya menampilkan kolom icon dimana isi itemnya didalamnya tidak terlalu banyak
##### dan hasilnya seperti di dalam

```python
data_pie = df["icon"].value_counts()
data_pie.head()
```

```python
data_pie = df["icon"].value_counts().rename_axis("icon")
data_pie.head()
```

```python
data_pie = df["icon"].value_counts().rename_axis("icon").reset_index(name="id_count")
data_pie.head()
```

```python
plt.title("Grafik Pie")
plt.pie(data_pie["id_count"],labels=data_pie["icon"],startangle=90,autopct="%.2f%%")
plt.show()
```
