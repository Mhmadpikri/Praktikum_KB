```python
import pandas as pd
import numpy as np

df =pd.read_csv('Weather.csv')
```

#### masukkan dalam dictionary 

```python
dict = {
    'attributes': 0,
    'null_attributes': 0,
    'object_attributes': 0,
    'total_records': 0,
    'number_attributes':0
}
```

```python
dict['attributes'] = list(df.columns.values)
dict['null_attributes'] = list(df.loc[:, df.isnull().any()].columns)
dict['object_attributes'] = list(df.select_dtypes(['object']).columns)
dict['total_records'] = df.shape[0]
dict['number_attributes'] = list(df.select_dtypes(['float64', 'int64']).columns)
```
